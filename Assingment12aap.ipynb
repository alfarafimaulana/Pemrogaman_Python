```
import numpy as np
import matplotlib.pyplot as plt
```

```
def cos(frek,tipe,ht,hr,d):
    if frek in range(150,1501):
        Ca=69.55
        Cb=26.16
    elif frek in range(1501,2001):
        Ca=46.3
        Cb=33.9
    
    if tipe=='urban' :
        Cm=0
    elif tipe=='suburban' :
        Cm= -2*((np.log10((frek/28)**2))-5.4)
    elif tipe=='open' :
        Cm= -4.78*(np.log10(frek))+18.33*(np.log10(frek))-40.94
    else :
        Cm='tipe salah'
    
    ahr = (1.1*np.log10(frek)-0.7)*hr-(1.56*np.log10(frek)-1.8)
    Lp = Ca+(Cb*np.log10(frek))-(13.83*np.log(ht))-ahr+((44.9-6.55*np.log10(ht))*np.log10(d))-Cm
    
    return Lp
```

```
f=1800 
loc1='urban'
loc2='suburban'
loc3='open'
tower=80 
user=1.5 
dmin=200 
dmax=700 
N=200 
d=np.random.randint(dmin,dmax,N)
```

```
LpUrban=cos(f,'urban',tower,user,d)
LpSuburban=cos(f,'suburban',tower,user,d)
LpOpen=cos(f,'open',tower,user,d)
```

```
plt.figure()
plt.plot(d,LpUrban)
plt.plot(d,LpSuburban)
plt.plot(d,LpOpen)
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```

```
d_sort=d[np.argsort(d)]
Urban_sort=LpUrban[np.argsort(d)]
Suburban_sort=LpSuburban[np.argsort(d)]
Open_sort=LpOpen[np.argsort(d)]
```

```
plt.figure()
plt.plot(d_sort,Urban_sort, 'k-o', label='Urban')
plt.plot(d_sort,Suburban_sort, 'g-x', label='Suburban')
plt.plot(d_sort,Open_sort, 'r--', label='Open')
plt.legend()
plt.title('Assignment 1 Urban, Suburban, Open')
plt.xlabel('Jarak (m)')
plt.ylabel('Path Loss (dB)')
plt.show()
```
